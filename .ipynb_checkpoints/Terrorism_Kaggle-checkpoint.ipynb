```python
# Dependencies 
import numpy as np
import pandas as pd
import plotly.offline as py
import csv 
import plotly.graph_objs as go
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading data
terror=pd.read_csv('../gtd/globalterrorismdb_0718dist.csv' ,encoding='ISO-8859-1')
print("Data before extraction to specific columns: ",terror.shape)

# Rename columns which will be used + add 
terror.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)

# Columns used
terror=terror[['Year','Month','Day','Country','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]

# Add column casualities
terror['casualities']=terror['Killed']+terror['Wounded']
```

```python
# Shape after 
print("Data after extraction to specific columns: ",terror.shape)
```

```python
# What kind of data do we have
terror.get_dtype_counts()
```

```python
print('Country with Highest Terrorist Attacks:',terror['Country'].value_counts().index[0])
print('Month with Highest Terrorist Attacks:',terror['Month'].value_counts().index[0])
print('Day with Highest Terrorist Attacks:',terror['Day'].value_counts().index[0])
print('Region with Highest Terrorist Attacks:',terror['Region'].value_counts().index[0])
```


#### Country - Iraq - Muslim
#### Month - 5 - Muslims starts Ramadan
#### Day - 15 - Most common day when the Ramadan start 


```python
Country_terror = terror['Country'].value_counts(dropna= False)

plt.subplots(figsize=(15,6))
sns.barplot(terror['Country'].value_counts()[:20].index, terror['Country'].value_counts()[:20].values,palette='RdBu')
plt.xticks(rotation=90)
plt.title('Top Affected Countries')
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=terror,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year around the world')
plt.show()
```
